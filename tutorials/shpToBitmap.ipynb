```python
import geopandas as gpd
from shapely.geometry import mapping
import matplotlib.pyplot as plt
```

```python
import os
import sys
module_path = os.path.abspath(os.path.join('../'))
if module_path not in sys.path:
    sys.path.append(module_path)
```

```python
shp_data = gpd.read_file(r"shpToBitmapDemo.geojson")
shp_data.plot()
```

```python
feature_geometry = list()

for geometry in list(mapping(shp_data["geometry"])["features"]):
    feature_geometry.append(geometry["geometry"])
```

```python
from gkit.imgops import bitmap_gen, geowrite
for i, bmap in enumerate(bitmap_gen(shp_data.total_bounds, output_image_size=(100, 100), pixel_resolution=(1, 1), geometry_collection=feature_geometry)):
    # geowrite(rf"{i}.tiff", bmap.array, bmap.transform)
    plt.figure()
    plt.imshow(bmap.array)
    plt.show()
```
